```python
# Set the percentage threshold
percentage_threshold = 3

# Function to filter rows based on top percentage
def filter_top_percentage(df, column_name, percentage_threshold):
    threshold_value = df[column_name].quantile(1 - percentage_threshold / 100)
    return df[df[column_name] <= threshold_value]

# Iterate through each column and filter rows
for column in df.columns:
    df = filter_top_percentage(df, column, percentage_threshold)
```
